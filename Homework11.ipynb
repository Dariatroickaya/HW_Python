```python
# Даны две функции: 
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# Найти точки пересечения
# Построить графики функций в одной системе координат
# Построить график общей функции
```

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
fx = x**3 - 50*x
gx = -1*x**4 + 88*x**2 - 241
fgx = fx - gx
```

```python
roots = solve(fgx)
roots
```

```python
count_roots = len(roots)
count_roots
```

```python
cross_dict = {}
i = 1
for j in roots:
    cross_dict[i] = {'x': round(N(j), 5), 'y': round(N(fx.evalf(subs={x: j}, chop=True)), 5)}
    i += 1
cross_dict
```

```python
print(*[(cross_dict[i]['x'], cross_dict[i]['y']) for i in cross_dict.keys()], sep=', ')
```
